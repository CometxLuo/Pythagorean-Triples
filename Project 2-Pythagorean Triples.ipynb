## Introduction

**Pythagorean Theorem:** 
The Pythagorean theorem, sometimes known as Pythagoras' theorem, is a basic relationship among the three sides of a right triangle in Euclidean geometry. With respect to this rule, the area of the square on the hypotenuse, the side opposite the right angle, $(c)$ equals to the sum of the areas of the two squares on the legs $(a$ and $b)$. This theorem can be expressed as the Pythagorean equation*, which is an equation connecting the lengths of the sides $a, b$ and the hypotenuse $c$.

$$^*a^2+b^2=c^2 $$

</div>


Along with the Pythagorean theorem, a triple of positive integers $(a, b, c)$ such that $a^2+b^2=c^2$ is called a *Pythagorean triple*. In such case, we call $(a,b)$ *Pythagorean tuple*.
        
**Example:** $(3, 4, 5)$ is a Pythagorean triple since $3^2+4^2=5^2$. Besides, $(3,4)$ is one set of Pythagorean tuples.

## Part 1. Find Pythagorean Triples

In order to find a list of all Pythagorean trples with the requiment, we need to write a function `ptriples(n)` that takes as its argument a number `n` and return a list of a triple of positive integers (a,b,c) where $1\leq a, b\leq n$. 

As the Pythagorean equation involved such operations as squares, square roots and exponents, we introduce the `math` module to facilitate the usage of functions easier.

```python
import math as m      # Import the math module and named as 'm' because we need the 'square root' and 'power' function.
```

To begin with, let's make sure that every number $a$ less than $n$ to pair with another number $b$ but the number $b$ should be every number less than $n$ as well. To achieve this step, we will set one `for loop` inside of the other `for loop`. Next step, each pair of $a$ and $b$ we got should be run to compute the square root of the sum of $a$ squared and $b$ squared. The computed result is named $c$. However, the chief form of the computed result is not an *integer* but a *float*. After checking to see if the number $c$ has the non-zero fractional part, we collect the $c$'s with zero fractional part into an initialized list. Utilizing the *Boolean values* can solve the checking problem. Set up an *equality test* (use `==` symbol) to comparise whether the computed result $c$ equals to the integer $c$. If the result $c$ has a non-zero fractional part, it won't be same with the integer $c$. Whereupon we convert the satisfied result $c$ to an integer. Along with every $a$ and $b$ being checked, every number $c$ which meets the requirement will be added to the list so that we can find out all Pythagorean triples we want. 

```python
def ptriples(n):                                # Start defining a function named 'ptriples'.
    list = []                                   # Initialize a list.
    
    for a in range(1, n + 1):                   # Set up a 'for loop' to run every number 'a' less than 'n' 
        for b in range(1, n + 1):               # Set up a 'for loop' to run every number 'b' less than 'n'
            c = m.sqrt(m.pow(a,2) + m.pow(b,2)) # Calculate 'c' by getting the square root of the sum of the 'a' squared and 'b' squared.
            if c == int(c):                     # Check that the computed result 'c' is a integer number or a number with non-zero fractional part.
                c = int(c)                      # Convert the float $c$ into integer form.
                list.append([a, b, c])          # Add the 'c' that meets the requirement to the list with the 'a' and 'b' which make the 'c', in term of (a, b, c).
            
    return list                                 # The loop ends then get the list of all Pythagorean triples.
```

Test if the `ptriples` function works by a known result.

```python
ptriples(20)          # Apply the 'ptriples' function by a known example.
```

## Part 2. Plot All Pythagorean Tuples

In the previous part, we are able to get several lists of Pythagorean triples by using the `ptriples` function. The first two elements of a Pythagorean triple, the $a$ and the $b$, can be combined to form a pair $(a,b)$ termed as the Pythagorean tuple. 

So as to observe the structure of the Pythagorean triples, we can plot several graphs of Pythagorean tuples with different ranges of `n` where $1\leq a, b\leq n$. It's one of the way to take stock of the pattern in them. Let `x` be the value $a$ and `y` be the value $b$.

Plotting graph requires a `matplotlib` package, specifically the `pyplot` module within the package.

```python
import matplotlib.pyplot as plt          # Import the 'matplotlib.pyplot' module, named it 'plt' for ease of use.
```

Create a function that accepts as its argument a number `n` and return a list of every first element $a$ of all Pythagorean triples $(a,b,c)$ where $1\leq a,b \leq n$. Use a `for loop` to extract every number $a$ as an $x$ value.

```python
def findx(n):                    # Start defining a function using to collect the 'x' value of 'a'.
    list_a = []                  # Initialize a list for later adding every 'a' less than 'n'.
    
    for triple in ptriples(n):   # Set up a 'for loop' to iterate every triple in the list of 'ptriples(n)'.
        list_a.append(triple[0]) # Add every first element of all triples in the list of 'ptriples(n)'.
    
    return list_a                # Return the list of collecting all 'a' values.
```

Test if the `findx` function works using a known result.

```python
findx(20)         # Apply the 'findx' function for 'n' = '20'.
```

Create a function that accepts as its argument a number `n` and return a list of every second element $b$ of all Pythagorean triples $(a,b,c)$ where $1\leq a,b \leq n$. Use a `for loop` to extract every number $b$ as a $y$ value.

```python
def findy(n):                    # Same idea as 'findx' function.       
    list_b = []
    
    for triple in ptriples(n):
        list_b.append(triple[1]) # Add every second element of all triples in the list of 'ptriples(n)'.
    
    return list_b
```

Test if the `findy` function works using a known result.

```python
findy(20)         # Apply the 'findy' function for 'n' = '20'.
```

On the purpose of plotting Pythagorean tuples with four different `n` values to compare, we are going to use the `plt.subplot` function to build a matplotlib figure from four graphs. Find the `x` and `y` values corresponding four different `n` values first. Then create eight lists of their `x` and `y` values. With the `plt.tight_layout` function applied, the four graphs will be adjusted to the proper spacing. 

For a more intuitive and aesthetical view of these diagram, we select the point marker `.`, give each graph one particular title by the `plt.title` function, and label their axises by the `plt.xlabel` and `plt.ylabel` function.

```python
x1 = findx(20)           # Let 'x1' be a list variable of all the 'x' values of the 'a', which is the first element of all triples in the list of 'ptriples(20)'.
y1 = findy(20)           # Similarly obtain the 'y' value. Repeat it for 'n' = '100, 500, 1000'.

x2 = findx(100)
y2 = findy(100)

x3 = findx(500)
y3 = findy(500)

x4 = findx(1000)
y4 = findy(1000)

plt.subplot(2,2,1)                         # Set up a matplotlib figure consisting of four subplots. 
                                           # It is a 2x2 layout grid. The graph locates at the first row, first column.
plt.plot(x1,y1,'.')                        # 'x' value adopts the first element of 'ptriples(20)'; 'y' value adopts the second element of 'ptriples(20)'; use the point marker.
plt.title('20 sets of Pythagorean Tuples') # Add a title using 'plt.title'
plt.xlabel('$a$')                          # Label the x axis 'a'.
plt.ylabel('$b$')                          # Label the y axis 'b'.
                       
                                           # Repeat operation for 'n' = '100, 500, 1000'.
plt.subplot(2,2,2)
plt.plot(x2,y2,'.')
plt.title('100 sets of Pythagorean Tuples')
plt.xlabel('$a$')
plt.ylabel('$b$')

plt.subplot(2,2,3)
plt.plot(x3,y3,'.',markersize = "1")
plt.title('500 set of Pythagorean Tuples')
plt.xlabel('$a$')
plt.ylabel('$b$')

plt.subplot(2,2,4)
plt.plot(x4,y4,'.',markersize = "1")
plt.title('1000 sets of Pythagorean Tuples')
plt.xlabel('$a$')
plt.ylabel('$b$')

plt.tight_layout()                          # Fix the spacing on its own.
```

## Part 3. Analyze the Structure of Pythagorean Tuples

It is apparent that the scatter plots we created in Part 2 have some distinct characteristics. Subsequently, we will seek out the causes for the occurrence of those properties for the purpose of investigating the structure of Pythagorean tuples.

**Description of observable characteristics:**

As the amount of the points increases, it becomes clear that those points will emerge eventually as some straight lines in the diagram. Examine closely, the two line in the deepest color, which means the most prominent two lines, are made up of a series of points $(3,4),(6,8),(9,12)...$ and another series of points $(4,3),(8,6),(12,9)...$. In other word, for every right triangle that can form a Pythagorean triple, lengthening three sides by the same integer multiple (absolutely positive) can generate a larger right triangle, which can also constitute a larger Pythagorean triple. 

For instance, the Pythagorean tuple produced by the two legs of a right triangle with a hypotenuse of length $10$, is $(6,8)$ or $(8,6)$. A triangle with three integer sides of $18,24,30$, is a similar triangle of the instance triangle. 

$$6\cdot 3=18$$
$$8\cdot 3=24$$
$$10\cdot 3=30$$
Same multiple: $3$
$$6^2+8^2=10^2$$
$$18^2+24^2=30^2$$

When three sides were stretched by the same multiple, it turns out a larger comparable right triangle. Put it another way, the reason we get some straight lines in the matplotlib figure is because there are thousands of right triangles that are similar to a primitive right triangle. Those tuples generated by similar triangles all have different integer multiples of the primitive tuple. As a result, the slopes made from the origin and the points of Pythagorean tuples are identical (slope = $\frac{b}{a}$). 

**Primitive Pythagorean Triples:** The Pythagorean triple is said to be **primitive** if and only if the greatest common divisor of $a,b,$ and $c$ is $1$. By contrast, $(3,4,5)$ is a primitive Pythagorean triple whereas $(6,8,10) is not.

Now, let's attempt to plot all the primitive Pythagorean tuple.

Write a fuction for finding $gcd(a,b)$.

1. Start with two integers $a$ and $b$.
2. Divide the integer $b$ by $a$. Replace the $a$ with $b$, and replace $b$ with the reminder. If $a$ is smaller than $b$ at the first time, $a$ % $b=a$. They will exchange the value in the first loop.
3. Repeat the $2^{nd}$ step until the $b$ equals to zero.
4. The $a$ will be the $gcd(a,b)$ while the loop ends.

```python
def gcd(a,b):             # Define a function for finding greatest common divisor
    while b != 0:         # Set up a while loop. Since we don't know when it will end, we choose 'b' = 0 to end the loop.
        a, b = b, a % b   # In order to let 'a' always be the largest number, replace 'a' with 'b' and replace 'b' with the reminder of 'a' divides 'b'.
    return a              # 
```

Test the `gcd` function.

```python
gcd(48,60)          # Apply 'gcd' function to integers 48 and 60.
```

Actually we got the `math.gcd` function. Let us check if we got the same result for these two 'gcd' functions.

```python
m.gcd(48,60)       # Apply 'math.gcd' function to integers 48 and 60.
```

The `gcd` function that we wrote works. Let us use it to find several primitive Pythagorean triples with $n=100$.

```python
ptriples_list = ptriples(3500)

pptriples = [(a, b, c) for (a, b, c) in ptriples_list if gcd(a,b) == 1]

pptriples
```

We don't want the repeated Pythagorean triples. The fact that repeated Pythagorean triples $(a, b, c)$ have things in common is that $a,b,$ and $c$ add up to the same amount. To achieve it, we can set a `for` loop to check which triples has the same sum with the previous checked ones. Then filter out the repeated.

```python
sum_of_abc = []                  # Initialize a list for adding the sum of 'a', 'b', and 'c' later.

new_list = []                    # Initialize a list for adding the non-repeated triples later.

for i in pptriples:              # Iterate each triples in 'pptriples'.
    sum = i[0] + i[1] + i[2]     # get the sum of 'a', 'b', and 'c'.
    if sum not in sum_of_abc:    # Check if the sum appears before.
        sum_of_abc.append(sum)   # Add the sum for reference.
        new_list.append(i)       # If the triple gets in the loop, that means it isn't repeated. Then we add it into the list for collecting nonrepeated triples.
        
new_list                         # Get the list of nonrepeated triples when the loop ends.
```

Plot the primitive Pythagorean tuples by using the `plt.plot` function.

```python
x = []                                               # Initialize a list for 'x' value.
y = []                                               # Initialize a list for 'y' value.

for triple in new_list:
    x.append(triple[0])                              # Add every first element of all triples in the 'new_list'.
    y.append(triple[1])                              # Add every second element of all triples in the 'new_list'.

plt.plot(x,y,'.')                                    # 'x' value adopts every first element of 'new_list'; 'y' value adopts every second element of 'new_list'. Use the point marker.
plt.title('Primitive Pythagorean Tuples')            # Add a title using 'plt.title'.
plt.xlabel('$a$')                                    # Label the x axis 'a'.
plt.ylabel('$b$')                                    # Label the y axis 'b'.
```
